```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as npl
```

```python
array = np.array([5,10])
array_2 = np.array([9,16])
print(array)
print()
print(array_2)
```

```python
def graficarVectores(vecs,cols):
  plt.figure(figsize=(5,5))
  for i in range(len(vecs)):
    vec=vecs[i]
    plt.quiver(0,0,vec[0],vec[1],color=cols[i],angles='xy',scale_units='xy',scale=1)
```

```python
graficarVectores([array,array_2],['blue','red'])
plt.xlim(0, max(array[0],array_2[0]))
plt.ylim(0, max(array[1],array_2[1]))
plt.title("Matriz")
plt.legend(["Vector 1","Vector 2"]);
plt.grid()
```

```python
P_punto = np.dot(array, array_2)
print(P_punto)
```

```python
long_1 = np.sqrt(np.dot(array,array))
print(long_1)
```

```python
long_2 = np.sqrt(np.dot(array_2,array_2))
print(long_2)
```

```python
cos_angle = lambda array, array_2:(P_punto/(long_1*long_2))
angle = lambda cos:np.arccos(cos)*(180/np.pi)
r=angle(cos_angle(array,array_2))
print(r)
```

```python
def gram_schmidt(array):
  tam = array.shape
  m=tam[0]
  n=tam[1]
  array_u=np.zeros((m,n))
  array_u[0] = array[0]

  for k in range(1,m):
    array_u[k]=array[k]
    for j in range(k):
      array_u[k]=array_u[k]-(np.dot(array[k],array_u[j]))/(np.dot(array_u[j],array_u[j]))*array_u[j]

  return array_u
gram_schmidt(array)
```

```python
b = np.vstack([5,3])
x =np.vstack([-2,7])
print(b)
print()
print(x)
```

```python
lambd = np.dot(b.T,x)/npl.norm(b)**2 # b^T x / ||b||^2;
print(lambd)
```

```python
vec_proy=b*lambd
print(vec_proy)
```

```python
matr_proy = np.dot(b,b.T)/npl.norm(b)**2#bb^T/||b||^2
print(matr_proy)
```

```python
p=np.dot(matr_proy,x)
print(p)
```

```python
a = np.array([[5,1],[2,2]])
print(a)
```

```python
res, res_2 = npl.eig(a)
print(res)
print()
print(res_2)
```

```python
autovector= npl.eig(a)[1][:,1]
autovector_2= npl.eig(a)[1][:,0]
print(autovector)
print(autovector_2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
scale = 10
origin2D = np.vstack([0,0])

plt.axis([-scale/2,scale/2,-scale/2,scale/2])
plt.quiver(*origin2D,*a[:,0],scale=scale, label='Vectores Originales')
plt.quiver(*origin2D,*a[:,1],scale=scale)

plt.quiver(*origin2D, *(autovector*5), scale=scale,color="gold", label='Autovectores')
plt.quiver(*origin2D, *(autovector_2*2), scale=scale,color="gold")

ax.set_aspect('equal', adjustable='box')
plt.legend();
```
